```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('IHME_GBD_2019_CHEWING_TOB_1990_2019_DATA_Y2021M05D27.CSV',encoding='latin1')
df
```

```python
df.dropna()
```

```python
ser = df['val']
ser
```

```python
# obliczanie Å›redniej

np.mean(ser)
```

```python
#obliczanie mediany 

np.median(ser)
```

```python
#Odchylenie standardowe

np.std(ser)
```

```python
#Wariancja 

np.var(ser)
```

```python
ser2 = df['lower']

ser2
```

```python
#Koleracja

correlation = np.corrcoef(ser,ser2) [0,1]
print(correlation)
```

```python
#Kowariancja 

covariance = np.cov(ser,ser2) [0,1]
covariance
```
